### 17. What is causation? Explain the difference between correlation and causation with an example.

Answer:

Causation means one event directly causes another event to happen.

Correlation means two variables move together or change together, but it does not mean that one is causing the other.

Example:
During summer, ice cream sales increase and sunburn cases also increase. These two are correlated because both are affected by hot weather. However, eating ice cream does not cause sunburn. The real cause is the heat from the sun.

### 18. What is an Optimizer? What are different types of optimizers? Explain each with an example.

Answer:

An optimizer is an algorithm used in machine learning and deep learning to update the model’s parameters (weights and biases) so that the loss or error becomes smaller during training.

Some common optimizers are:

SGD (Stochastic Gradient Descent):
It updates the model parameters using one training example at a time. It is simple but can be slower in convergence.

```python
from tensorflow.keras.optimizers import SGD

optimizer = SGD(learning_rate=0.01)
```

Adam (Adaptive Moment Estimation):
Adam automatically adjusts the learning rate during training and combines the advantages of two other methods. It is one of the most commonly used optimizers.

```python
from tensorflow.keras.optimizers import Adam

optimizer = Adam(learning_rate=0.001)
```

RMSprop:
RMSprop changes the learning rate based on the average of recent gradients, which helps make the training process more stable.

```python
from tensorflow.keras.optimizers import RMSprop

optimizer = RMSprop(learning_rate=0.001)
```
