# Q11 Identify and remove outliers in km_driven using IQR method

```python
Q1 = df["km_driven"].quantile(0.25)
Q3 = df["km_driven"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df_clean = df[(df["km_driven"] >= lower) & (df["km_driven"] <= upper)]

df_clean.shape
```

# EDA-4 : Exploratory Data Analysis (EDA) on Spotify Data: Popular Hip-hop Artists and Tracks

## Q1. Read the dataframe, check for null values and duplicate rows, and handle them if present.

```python
import pandas as pd

df = pd.read_csv("spotify_hiphop_tracks.csv")

# check null values
df.isnull().sum()

# handle null values
df = df.dropna()

# check duplicates
df.duplicated().sum()

# remove duplicates
df = df.drop_duplicates()

df.head()
```

---

## Q2. What is the distribution of popularity among the tracks in the dataset?

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df["Popularity"], bins=20, kde=True)

plt.title("Distribution of Track Popularity")
plt.xlabel("Popularity")
plt.ylabel("Count")

plt.show()
```

---

## Q3. Is there any relationship between popularity and the duration of tracks?

```python
sns.scatterplot(x="Duration (ms)", y="Popularity", data=df)

plt.title("Popularity vs Track Duration")

plt.show()
```

---

## Q4. Which artist has the highest number of tracks in the dataset?

```python
sns.countplot(y="Artist", data=df, order=df["Artist"].value_counts().index)

plt.title("Number of Tracks per Artist")

plt.show()
```

```python
df["Artist"].value_counts()
```

---

## Q5. What are the top 5 least popular tracks in the dataset?

```python
df.nsmallest(5, "Popularity")[["Artist", "Track Name", "Popularity"]]
```

---

## Q6. Among the top 5 most popular artists, which artist has the highest average popularity?

```python
top_artists = df.groupby("Artist")["Popularity"].mean().sort_values(ascending=False).head(5)

top_artists
```

---

## Q7. For the top 5 most popular artists, what are their most popular tracks?

```python
top_artist_names = top_artists.index

df[df["Artist"].isin(top_artist_names)].sort_values("Popularity", ascending=False)[["Artist","Track Name","Popularity"]]
```

---

## Q8. Visualize relationships between multiple numerical variables using a pair plot.

```python
sns.pairplot(df[["Popularity","Duration (ms)"]])

plt.show()
```

---

## Q9. Does the duration of tracks vary significantly across different artists?

```python
sns.boxplot(x="Artist", y="Duration (ms)", data=df)

plt.xticks(rotation=90)

plt.title("Track Duration Across Artists")

plt.show()
```

---

## Q10. How does the distribution of track popularity vary for different artists?
