Q1) What is hypothesis testing in statistics.

Hypothesis testing is a method used in statistics to test a hypothesis or statement about a population parameter, based on sample data. It involves comparing observed data to what we would expect under the null hypothesis..


Q2 What is the null hypothesis, and how does it differ from the alternative hypothesis?


Null Hypothesis (H0) - A statement that there is no effect or no difference,
and it assumes that any observed effect in the data is due to random chance.

Alternative Hypothesis (H1) - A statement that contradicts the null hypothesis, indicating that there is an effect or a difference.


Q3 What is the significance level in hypothesis testing, and why is it important*

 Significance level (alpha) is the probability of rejecting the null hypothesis when it is actually true (Type I error). It is important because it sets the threshold for determining whether the observed data are statistically significant.



Q4 What does a P-value represent in hypothesis testing*

A P-value represents the probability of obtaining the observed data, or more extreme, assuming the null hypothesis is true.


Q5 How do you interpret the P-value in hypothesis testing*

 If the P-value is less than the significance level (alpha), we reject the null hypothesis. If the P-value is greater than the significance level, we fail to reject the null hypothesis.

Q6  What are Type 1 and Type 2 errors in hypothesis testing*

Type 1 Error - Occurs when the null hypothesis is true, but we incorrectly reject it (false positive).

Type 2 Error - Occurs when the null hypothesis is false, but we fail to reject it (false negative).



Q7 What is the difference between a one-tailed and a two-tailed test in hypothesis testing*

One-tailed Test - Tests if the sample mean is significantly greater than or less than the population mean. Used when the alternative hypothesis is directional.

Two-tailed Test - Tests if the sample mean is significantly different from the population mean in either direction. Used when the alternative hypothesis is non-directional.


Q8  What is the Z-test, and when is it used in hypothesis testing*

If our sample size is greater than 30 and the standard deviation of the population is given, we use the Z-test. It is used in hypothesis testing for finding the mean difference between the population and the sample.

Q9 How do you calculate the Z-score, and what does it represent in hypothesis testing*

To calculate the Z-score, divide the difference between the sample mean and the population mean by the standard error. The Z-score represents the number of standard deviations the sample mean is from the population mean in hypothesis testing.

Q10 What is the T-distribution, and when should it be used instead of the normal distribution*

The T-distribution is used to find the difference between the population mean and the sample mean. If we have a sample size less than 30 and the population standard deviation is unknown, we use the T-test/T-distribution. It is close to the standard normal distribution.

Sample Size: In the Z-test, the sample size is greater than 30, while in the T-test, the sample size is less than 30.

Q12 What is the T-test, and how is it used in hypothesis testing*

If the sample size is less than 30 and the population standard deviation is unknown, we use the T-test. It is used in hypothesis testing to find the mean difference between the population and the sample mean.


Q13 What is the relationship between Z-test and T-test in hypothesis testing*

Both are used to find the mean difference between the population and the sample mean.


Q14 What is a confidence interval, and how is it used to interpret statistical results*

A confidence interval is the range within which we can say that a certain percentage of our data lies. To find the confidence interval, we use the T-test or Z-test. It helps in interpreting statistical results by providing a range that is likely to contain the population parameter.

Q15 What is the margin of error, and how does it affect the confidence interval

The margin of error is the value that indicates the range within which our hypothesis testing is likely to be accurate. If the margin of error is high, it means the range of the confidence interval is wide.

The Chi-square distribution is used to find the variance difference in categorical variables. It has only one parameter, the degree of freedom, and is always right-skewed. If the degree of freedom increases, it becomes more similar to the normal distribution. It is used in the Chi-square test to determine if there is a significant difference between population variance and sample variance.

The F-distribution is used when we have more than three variables and we want to find the mean difference among these variables. It is used in the ANOVA test and regression analysis.

Q4 Perform a two-tailed Z-test using Python and visualize the decision region on a plot

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Perform a two-tailed Z-test for a given sample mean against a known population mean and visualize the decision region.

    Parameters:
    sample (list or array): Sample data
    pop_mean (float): Known population mean
    pop_std (float): Known population standard deviation
    alpha (float): Significance level (default: 0.05)

    Returns:
    None
    """
    # Sample statistics
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Compute Z-score
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

    # Compute p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Print results
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Decision rule
    z_critical = stats.norm.ppf(1 - alpha / 2)  # Two-tailed critical value

    # Visualizing the decision regions
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x, 0, 1)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label="Standard Normal Distribution")
    plt.axvline(-z_critical, color='red', linestyle='dashed', label=f"Critical Value -{z_critical:.2f}")
    plt.axvline(z_critical, color='red', linestyle='dashed', label=f"Critical Value {z_critical:.2f}")
    plt.axvline(z_score, color='blue', linestyle='solid', label=f"Z-score {z_score:.2f}")
    plt.fill_between(x, y, where=(x < -z_critical) | (x > z_critical), color='red', alpha=0.3, label="Rejection Region")

    plt.legend()
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.title("Two-tailed Z-test Decision Regions")
    plt.show()

    # Interpret the result
    if abs(z_score) > z_critical:
        print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis: No significant difference between sample and population means.")

# Simulate random sample data
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=4, size=30)  # Normal distribution with mean 50 and std 4

# Population parameters
population_mean = 50
population_std = 4

# Perform the Z-test and visualize
test_result = z_test(sample_data, population_mean, population_std)
```

Q5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Perform a two-tailed Z-test for a given sample mean against a known population mean and visualize the decision region.
    """
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    z_critical = stats.norm.ppf(1 - alpha / 2)

    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    if abs(z_score) > z_critical:
        print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis: No significant difference between sample and population means.")

def visualize_type1_error(alpha=0.05):
    """Visualize Type 1 error (False Positive)."""
    x = np.linspace(-4, 4, 1000)
    y_null = stats.norm.pdf(x, 0, 1)
    z_critical = stats.norm.ppf(1 - alpha)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y_null, label="Null Hypothesis (H0)", color='blue')
    plt.fill_between(x, y_null, where=(x > z_critical), color='red', alpha=0.3, label="Type 1 Error (Î±)")
    plt.axvline(z_critical, color='black', linestyle='dashed', label=f"Critical Value {z_critical:.2f}")

    plt.legend()
    plt.xlabel("Test Statistic")
    plt.ylabel("Probability Density")
    plt.title("Type 1 Error in Hypothesis Testing")
    plt.show()

def visualize_type2_error(beta=0.2):
    """Visualize Type 2 error (False Negative)."""
    x = np.linspace(-4, 4, 1000)
    y_alt = stats.norm.pdf(x, 2, 1)
    z_critical = stats.norm.ppf(1 - beta)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y_alt, label="Alternative Hypothesis (H1)", color='green')
    plt.fill_between(x, y_alt, where=(x < z_critical), color='purple', alpha=0.3, label="Type 2 Error (Î²)")
    plt.axvline(z_critical, color='black', linestyle='dashed', label=f"Critical Value {z_critical:.2f}")

    plt.legend()
    plt.xlabel("Test Statistic")
    plt.ylabel("Probability Density")
    plt.title("Type 2 Error in Hypothesis Testing")
    plt.show()

np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=4, size=30)

population_mean = 50
population_std = 4

test_result = z_test(sample_data, population_mean, population_std)
visualize_type1_error()
visualize_type2_error()
```

Q12 Perform a Chi-square test for independence between two categorical variables in Python

```python
import pandas as pd
import scipy.stats as stats

#sample data : two categorical variable (gender and preference)

data = {'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
        'Preference': ['A', 'A', 'B', 'B', 'A', 'B', 'B', 'A']}

#create a dataframe
df =pd.DataFrame(data)

#create a contingency table
contingency_table = pd.crosstab(df['Gender'] , df['Preference'])

#perform chi square test
chi2 , p , dof , expeceted = stats.chi2_contingency(contingency_table)
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expeceted)

alpha = 0.05
if p < alpha:
    print("Result: We reject the null hypothesis. There is a significant association between Gender and Preference.")
else:
    print("Result: We fail to reject the null hypothesis. There is no significant association between Gender and Preference.")
```

Q13 Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data

```python
import numpy as np
import pandas as pd

# Example: Observed frequencies in a contingency table
# Rows could represent different groups and columns could represent different categories.
# For example, we have 3 rows and 3 columns.
observed = np.array([
    [10, 20, 30],
    [20, 15, 25],
    [25, 30, 20]
])

# Calculate the totals for each row and each column, and the overall total.
row_totals = observed.sum(axis=1, keepdims=True)  # Sum along columns for each row
col_totals = observed.sum(axis=0, keepdims=True)    # Sum along rows for each column
grand_total = observed.sum()

# Calculate expected frequencies using the formula:
# Expected_ij = (Row Total_i * Column Total_j) / Grand Total
expected = (row_totals @ col_totals) / grand_total  # Using matrix multiplication (@)

# Display the results
print("Observed Frequencies:")
print(observed)
print("\nRow Totals:", row_totals.flatten())
print("Column Totals:", col_totals.flatten())
print("Grand Total:", grand_total)
print("\nExpected Frequencies:")
print(expected)
```

Q16 Implement an F-test using Python to compare the variances of two random samples

```python
import numpy as np
import scipy.stats as stats

# Step 1: Generate Random Samples
np.random.seed(42)
sample1 = np.random.normal(loc=10, scale=2, size=100)
sample2 = np.random.normal(loc=12, scale=3, size=100)

# Step 2: Calculate variances of the two samples
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Calculate the F-statistic
F_stat = var1 / var2

# Step 3: Perform F-test
df1 = len(sample1) - 1  # Degrees of freedom for sample 1
df2 = len(sample2) - 1  # Degrees of freedom for sample 2
p_value = stats.f.cdf(F_stat, df1, df2)  # One-tailed test

# To get the two-tailed p-value
p_value = 2 * min(p_value, 1 - p_value)

# Display the results
print("F-test Results:")
print(f"Sample 1 Variance: {var1:.4f}")
print(f"Sample 2 Variance: {var2:.4f}")
print(f"F-statistic: {F_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Result: We reject the null hypothesis. The variances of the two samples are significantly different.")
else:
    print("Result: We fail to reject the null hypothesis. There is no significant difference between the variances of the two samples.")
```

Q17 Write a Python program to perform an ANOVA test to compare means between multiple groups and
interpret the results

```python
import numpy as np
import scipy.stats as stats

# Step 1: Generate Sample Data
np.random.seed(42)
group1 = np.random.normal(loc=20, scale=5, size=30)
group2 = np.random.normal(loc=22, scale=5, size=30)
group3 = np.random.normal(loc=19, scale=5, size=30)

# Step 2: Perform ANOVA Test
f_stat, p_value = stats.f_oneway(group1, group2, group3)

# Step 3: Display and Interpret the Results
print("ANOVA Test Results:")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Result: We reject the null hypothesis. There is a significant difference between the group means.")
else:
    print("Result: We fail to reject the null hypothesis. There is no significant difference between the group means.")
```

Q19 Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(*groups):
    """
    Check the assumptions (normality, independence, and equal variance) for ANOVA.

    Parameters:
    groups: Variable number of group arrays (each group as a separate argument).

    Returns:
    None
    """

    # Check Normality using Shapiro-Wilk test
    print("Normality Check using Shapiro-Wilk Test:")
    for i, group in enumerate(groups, start=1):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i}: W = {stat:.4f}, p-value = {p_value:.4f}")
        if p_value < 0.05:
            print(f"Group {i}: Data is not normally distributed.")
        else:
            print(f"Group {i}: Data is normally distributed.")

    # Visualize Normality using Q-Q plots
    plt.figure(figsize=(15, 6))
    for i, group in enumerate(groups, start=1):
        plt.subplot(1, len(groups), i)
        stats.probplot(group, dist="norm", plot=plt)
        plt.title(f'Q-Q Plot Group {i}')
    plt.tight_layout()
    plt.show()

    # Check Homogeneity of Variance using Levene's test
    stat, p_value = stats.levene(*groups)
    print("\nHomogeneity of Variance Check using Levene's Test:")
    print(f"Levene's Test: W = {stat:.4f}, p-value = {p_value:.4f}")
    if p_value < 0.05:
        print("Variances are not equal across groups.")
    else:
        print("Variances are equal across groups.")

    # Visualize Homogeneity of Variance using box plots
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=groups)
    plt.title('Box Plot of Groups')
    plt.xlabel('Group')
    plt.ylabel('Values')
    plt.show()

    # Independence is assumed to be satisfied based on study design
    print("\nIndependence: Assumed to be satisfied based on study design.")

# Example usage
np.random.seed(42)
group1 = np.random.normal(loc=20, scale=5, size=30)
group2 = np.random.normal(loc=22, scale=5, size=30)
group3 = np.random.normal(loc=19, scale=5, size=30)

check_anova_assumptions(group1, group2, group3)
```

Q20 Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the
results

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as ols
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate Sample Data
np.random.seed(42)
factor1 = np.random.choice(['A', 'B'], size=60)
factor2 = np.random.choice(['X', 'Y'], size=60)
dependent_variable = np.random.normal(loc=10, scale=3, size=60) + \
                     (factor1 == 'B') * 2 + \
                     (factor2 == 'Y') * -1 + \
                     ((factor1 == 'B') & (factor2 == 'Y')) * 3

data = pd.DataFrame({'Factor1': factor1, 'Factor2': factor2, 'DependentVariable': dependent_variable})

# Step 2: Perform Two-Way ANOVA Test
model = ols.ols('DependentVariable ~ Factor1 * Factor2', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Step 3: Display and Interpret the Results
print("Two-Way ANOVA Test Results:")
print(anova_table)

# Check significance levels
alpha = 0.05
print("\nSignificance Results:")
for factor, row in anova_table.iterrows():
    print(f"{factor}: {'Significant' if row['PR(>F)'] < alpha else 'Not Significant'}")

# Step 4: Visualize the Results

# Interaction plot
interaction_plot = sns.pointplot(x='Factor1', y='DependentVariable', hue='Factor2', data=data, dodge=True, markers=['o', 's'], capsize=.1)
interaction_plot.set(title='Interaction Plot')
plt.show()

# Box plots
plt.figure(figsize=(10, 6))
sns.boxplot(x='Factor1', y='DependentVariable', hue='Factor2', data=data)
plt.title('Box Plot of Factors')
plt.xlabel('Factor1')
plt.ylabel('Dependent Variable')
plt.show()
```

Q21 Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def visualize_f_distribution(df1, df2):
    """
    Visualize the F-distribution with given degrees of freedom.

    Parameters:
    df1 (int): Degrees of freedom for the numerator.
    df2 (int): Degrees of freedom for the denominator.

    Returns:
    None
    """
    # Generate x values
    x = np.linspace(0, 5, 500)

    # Compute the F-distribution PDF
    f_pdf = stats.f.pdf(x, df1, df2)

    # Plot the F-distribution
    plt.figure(figsize=(10, 6))
    plt.plot(x, f_pdf, 'r-', lw=2, label=f'F-distribution (df1={df1}, df2={df2})')
    plt.fill_between(x, 0, f_pdf, color='red', alpha=0.3)

    # Add labels and title
    plt.xlabel('F-value')
    plt.ylabel('Density')
    plt.title(f'F-distribution with df1={df1} and df2={df2}')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
visualize_f_distribution(df1=5, df2=10)
```

Q22 Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate Sample Data
np.random.seed(42)
group1 = np.random.normal(loc=20, scale=5, size=30)
group2 = np.random.normal(loc=22, scale=5, size=30)
group3 = np.random.normal(loc=19, scale=5, size=30)
group4 = np.random.normal(loc=21, scale=5, size=30)

# Combine data into a DataFrame for visualization
data = pd.DataFrame({
    'Values': np.concatenate([group1, group2, group3, group4]),
    'Group': ['Group 1']*30 + ['Group 2']*30 + ['Group 3']*30 + ['Group 4']*30
})

# Step 2: Perform ANOVA Test
f_stat, p_value = stats.f_oneway(group1, group2, group3, group4)

# Step 3: Display and Interpret the Results
print("ANOVA Test Results:")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Result: We reject the null hypothesis. There is a significant difference between the group means.")
else:
    print("Result: We fail to reject the null hypothesis. There is no significant difference between the group means.")

# Step 4: Visualize the Results
plt.figure(figsize=(10, 6))
sns.boxplot(x='Group', y='Values', data=data)
plt.title('Box Plot of Group Means')
plt.xlabel('Group')
plt.ylabel('Values')
plt.show()
```

Q24  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

```python
import scipy.stats as stats

# Sample data
sample_variance = 20
sample_size = 15
population_variance = 25
alpha = 0.05

# Test statistic
chi_square_statistic = (sample_size - 1) * sample_variance / population_variance

# Degrees of freedom
degrees_of_freedom = sample_size - 1

# Critical values
chi_square_critical_low = stats.chi2.ppf(alpha / 2, degrees_of_freedom)
chi_square_critical_high = stats.chi2.ppf(1 - alpha / 2, degrees_of_freedom)

# Decision
if chi_square_statistic < chi_square_critical_low or chi_square_statistic > chi_square_critical_high:
    decision = "Reject the null hypothesis"
else:
    decision = "Fail to reject the null hypothesis"

# Output results
print(f"Chi-square statistic: {chi_square_statistic:.2f}")
print(f"Critical values: {chi_square_critical_low:.2f}, {chi_square_critical_high:.2f}")
print(f"Decision: {decision}")
```

Q25 Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

```python
import numpy as np
import scipy.stats as stats

def z_test_proportions(count1, nobs1, count2, nobs2, alpha=0.05):
    # Proportions
    prop1 = count1 / nobs1
    prop2 = count2 / nobs2

    # Pooled proportion
    pooled_prop = (count1 + count2) / (nobs1 + nobs2)

    # Standard error
    se = np.sqrt(pooled_prop * (1 - pooled_prop) * (1/nobs1 + 1/nobs2))

    # Z statistic
    z_stat = (prop1 - prop2) / se

    # P-value
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

    # Critical value
    z_critical = stats.norm.ppf(1 - alpha / 2)

    # Decision
    if abs(z_stat) > z_critical:
        decision = "Reject the null hypothesis"
    else:
        decision = "Fail to reject the null hypothesis"

    return {
        "z_stat": z_stat,
        "p_value": p_value,
        "z_critical": z_critical,
        "decision": decision
    }

# Example usage
count1 = 50  # Number of successes in group 1
nobs1 = 200  # Number of observations in group 1
count2 = 30  # Number of successes in group 2
nobs2 = 180  # Number of observations in group 2
alpha = 0.05

result = z_test_proportions(count1, nobs1, count2, nobs2, alpha)

print(f"Z statistic: {result['z_stat']:.2f}")
print(f"P-value: {result['p_value']:.4f}")
print(f"Critical value: {result['z_critical']:.2f}")
print(f"Decision: {result['decision']}")
```

Q26 Implement an F-test for comparing the variances of two datasets, then interpret and visualize the result

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generate two sample datasets
np.random.seed(42)
data1 = np.random.normal(loc=20, scale=5, size=100)
data2 = np.random.normal(loc=20, scale=10, size=100)

# Calculate sample variances
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# Perform F-test
f_stat = var1 / var2
df1 = len(data1) - 1
df2 = len(data2) - 1
p_value = stats.f.cdf(f_stat, df1, df2) if f_stat < 1 else 1 - stats.f.cdf(1 / f_stat, df2, df1)

# Determine critical value
alpha = 0.05
f_critical_low = stats.f.ppf(alpha / 2, df1, df2)
f_critical_high = stats.f.ppf(1 - alpha / 2, df1, df2)

# Decision
if f_stat < f_critical_low or f_stat > f_critical_high:
    decision = "Reject the null hypothesis"
else:
    decision = "Fail to reject the null hypothesis"

# Output results
print(f"F statistic: {f_stat:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Critical values: {f_critical_low:.2f}, {f_critical_high:.2f}")
print(f"Decision: {decision}")

# Visualize the data
plt.boxplot([data1, data2], labels=['Data1', 'Data2'])
plt.title('Boxplot of Two Datasets')
plt.ylabel('Values')
plt.show()
```
