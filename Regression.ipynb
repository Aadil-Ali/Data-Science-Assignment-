# Regression - Assignment Questions

## Q1. What is Simple Linear Regression?

Simple Linear Regression is a statistical method used to model the relationship between one independent variable (X) and one dependent variable (Y). It fits a straight line to the data to predict the value of Y based on X.

Equation:  
Y = mX + c

Where:  
m = slope of the line  
c = intercept

---

## Q2. What are the key assumptions of Simple Linear Regression?

1. Linearity – There should be a linear relationship between X and Y.  
2. Independence – Observations should be independent of each other.  
3. Homoscedasticity – Constant variance of residuals.  
4. Normality – Residuals should be normally distributed.  
5. No significant outliers.

---

## Q3. What does the coefficient m represent in the equation Y = mX + c?

The coefficient **m** represents the slope of the regression line. It indicates how much the dependent variable (Y) changes when the independent variable (X) increases by one unit.

---

## Q4. What does the intercept c represent in the equation Y = mX + c?

The intercept **c** represents the value of Y when X = 0. It is the point where the regression line crosses the Y-axis.

---

## Q5. How do we calculate the slope m in Simple Linear Regression?

The slope is calculated using the formula:

m = Σ((X - X̄)(Y - Ȳ)) / Σ((X - X̄)²)

Where:  
X̄ = mean of X  
Ȳ = mean of Y

---

## Q6. What is the purpose of the least squares method in Simple Linear Regression?

The least squares method is used to find the best fitting line by minimizing the sum of squared differences between the actual values and predicted values.

---

## Q7. How is the coefficient of determination (R²) interpreted in Simple Linear Regression?

R² measures how well the regression model explains the variability of the dependent variable.

Example:  
R² = 0.80 means 80% of the variation in Y is explained by X.

---

## Q8. What is Multiple Linear Regression?

Multiple Linear Regression is a regression technique used when the dependent variable depends on more than one independent variable.

Equation:

Y = b0 + b1X1 + b2X2 + ... + bnXn

---

## Q9. What is the main difference between Simple and Multiple Linear Regression?

Simple Linear Regression uses only one independent variable.  
Multiple Linear Regression uses two or more independent variables to predict the dependent variable.

---

## Q10. What are the key assumptions of Multiple Linear Regression?

1. Linearity  
2. Independence of errors  
3. Homoscedasticity  
4. No multicollinearity  
5. Normal distribution of residuals

---

## Q11. What is heteroscedasticity and how does it affect regression results?

Heteroscedasticity occurs when the variance of residuals is not constant across all levels of the independent variables. It can make the model unreliable and affect the accuracy of statistical tests.

---

## Q12. How can you improve a Multiple Linear Regression model with high multicollinearity?

Methods include:

1. Removing highly correlated variables  
2. Using Principal Component Analysis (PCA)  
3. Applying Regularization techniques such as Ridge or Lasso regression

---

## Q13. What are common techniques for transforming categorical variables in regression models?

1. Label Encoding  
2. One Hot Encoding  
3. Dummy Variables

---

## Q14. What is the role of interaction terms in Multiple Linear Regression?

Interaction terms allow the effect of one independent variable to depend on another independent variable.

Example:  
Y = b0 + b1X1 + b2X2 + b3(X1 × X2)

---

## Q15. How can the interpretation of intercept differ between Simple and Multiple Linear Regression?

In Simple Linear Regression, the intercept represents the value of Y when X = 0.

In Multiple Linear Regression, the intercept represents the value of Y when all independent variables are zero.

---

## Q16. What is the significance of the slope in regression analysis?

The slope indicates how much the dependent variable changes for a one-unit increase in the independent variable.

---

## Q17. How does the intercept provide context for the relationship between variables?

The intercept provides a baseline value of the dependent variable before the influence of independent variables.

---

## Q18. What are the limitations of using R² as a sole measure of model performance?

1. It does not indicate whether the model is appropriate.  
2. It always increases when more variables are added.  
3. It does not detect overfitting.  
4. It does not show prediction accuracy.

---

## Q19. How would you interpret a large standard error for a regression coefficient?

A large standard error indicates that the estimate of the coefficient is less reliable and there is high variability in the data.

---

## Q20. How can heteroscedasticity be identified in residual plots?

If residuals spread out in a funnel shape instead of forming a constant band around zero, it indicates heteroscedasticity.

---

## Q21. What does it mean if a regression model has high R² but low adjusted R²?

It means that some variables in the model are not useful and are increasing the R² artificially. Adjusted R² penalizes unnecessary variables.

---

## Q22. Why is it important to scale variables in Multiple Linear Regression?

Scaling ensures that variables with larger values do not dominate the model and helps improve numerical stability.

---

## Q23. What is Polynomial Regression?

Polynomial Regression is an extension of linear regression where the relationship between variables is modeled as a polynomial equation.

---

## Q24. How does Polynomial Regression differ from Linear Regression?

Linear regression fits a straight line, while polynomial regression fits a curved line to capture nonlinear relationships.

---

## Q25. When is Polynomial Regression used?

It is used when the relationship between independent and dependent variables is nonlinear.

---

## Q26. What is the general equation for Polynomial Regression?

Y = b0 + b1X + b2X² + b3X³ + ... + bnXⁿ

---

## Q27. Can polynomial regression be applied to multiple variables?

Yes, polynomial regression can be extended to multiple variables by including polynomial terms of different independent variables.

---

## Q28. What are the limitations of Polynomial Regression?

1. Risk of overfitting  
2. High complexity with higher degree polynomials  
3. Difficult interpretation

---

## Q29. What methods can be used to evaluate model fit when selecting polynomial degree?

1. Cross-validation  
2. Adjusted R²  
3. Mean Squared Error (MSE)  
4. Root Mean Squared Error (RMSE)

---

## Q30. Why is visualization important in Polynomial Regression?

Visualization helps to understand how well the polynomial curve fits the data and whether the model captures the pattern correctly.
